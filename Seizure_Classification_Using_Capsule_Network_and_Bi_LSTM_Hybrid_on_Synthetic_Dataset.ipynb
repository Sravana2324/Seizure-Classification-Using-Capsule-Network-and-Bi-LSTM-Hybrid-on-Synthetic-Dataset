```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import shap

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.utils.class_weight import compute_class_weight
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Input, Conv1D, MaxPooling1D, Dropout, LSTM, Bidirectional, Dense, Flatten, BatchNormalization, GRU
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from keras.regularizers import l2

# Custom Capsule Layer Class
def create_capsule_layer():
    # Define a custom Capsule Layer class
    class CapsuleLayer(tf.keras.layers.Layer):
        def __init__(self, num_capsules, capsule_dim, num_routing=3, **kwargs):
            # Initialize the Capsule Layer with the number of capsules, their dimensions, and the routing iterations
            super(CapsuleLayer, self).__init__(**kwargs)
            self.num_capsules = num_capsules
            self.capsule_dim = capsule_dim
            self.num_routing = num_routing

        def build(self, input_shape):
            # Build the layer by defining weights for the transformation matrix
            input_dim = input_shape[-1]  # Get the dimension of the input features
            self.W = self.add_weight(shape=[input_dim, self.num_capsules * self.capsule_dim],  # Transformation weights
                                     initializer='glorot_uniform',  # Xavier initialization
                                     trainable=True)  # Allow weights to be trainable

        def call(self, inputs):
            # Perform the forward pass
            inputs_reshaped = tf.nn.relu(inputs)  # Apply ReLU activation to the inputs
            # Calculate predictions (u_hat) by performing a dot product between inputs and transformation weights
            u_hat = tf.tensordot(inputs_reshaped, self.W, axes=[[2], [0]])
            # Reshape u_hat to match the number of capsules and capsule dimensions
            u_hat = tf.reshape(u_hat, (-1, inputs.shape[1], self.num_capsules, self.capsule_dim))

            # Initialize routing logits (b) with zeros
            b = tf.zeros_like(u_hat[:, :, :, 0])

            # Perform dynamic routing for the specified number of iterations
            for i in range(self.num_routing):
                # Compute coupling coefficients (c) using softmax across the capsule dimension
                c = tf.nn.softmax(b, axis=2)
                # Compute weighted sum of predictions (s) based on coupling coefficients
                s = tf.reduce_sum(c[..., None] * u_hat, axis=1)
                # Apply the squash function to get the output capsule vectors (v)
                v = self.squash(s)
                # Update routing logits (b) for the next iteration (if not the last iteration)
                if i < self.num_routing - 1:
                    b += tf.reduce_sum(u_hat * v[:, None, :, :], axis=-1)

            return v  # Return the final capsule outputs

        def squash(self, s, axis=-1):
            # Squash function to ensure output vectors have a length between 0 and 1
            s_squared_norm = tf.reduce_sum(tf.square(s), axis=axis, keepdims=True)  # Compute squared norm of s
            scale = s_squared_norm / (1 + s_squared_norm)  # Compute scale factor
            return scale * s / tf.sqrt(s_squared_norm + 1e-9)  # Apply squash formula

        def compute_output_shape(self, input_shape):
            # Define the output shape of the Capsule Layer
            return (input_shape[0], self.num_capsules, self.capsule_dim)

    return CapsuleLayer  # Return the Capsule Layer class

# Data Preprocessing
def load_and_preprocess_new_dataset(base_path):
    # Mapping class folder names to numeric labels
    class_mapping = {'Normal': 0, 'Focal': 1, 'General': 2}
    data, labels = [], []

    # Iterate through all folders in the base path
    for folder in os.listdir(base_path):
        folder_path = os.path.join(base_path, folder)
        print(f"\nProcessing folder: {folder}\n")

        # Skip folders not in the class mapping
        if folder not in class_mapping:
            continue

        file_count = 0  # Initialize file count for the current folder

        # Iterate through all files in the folder
        for file in os.listdir(folder_path):
            file_path = os.path.join(folder_path, file)
            print(f"\tReading file: {file}")
            try:
                # Load the CSV file as a DataFrame
                df = pd.read_csv(file_path)

                # Extract EEG data columns (first 25 columns excluding time column)
                eeg_data = df.iloc[:, 1:26].apply(pd.to_numeric, errors='coerce').dropna().values

                # Extract time column (optional, first column)
                time_data = df.iloc[:, 0].apply(pd.to_numeric, errors='coerce').dropna().values

                # Assign the numeric label for this folder
                label = class_mapping[folder]

                # Segment the data into overlapping 1-second chunks (178 data points per second per electrode)
                num_samples = len(eeg_data)
                for start in range(0, num_samples - 178, 89):  # 50% overlap between chunks
                    chunk = eeg_data[start:start + 178, :]

                    # Ensure the chunk contains exactly 178 samples
                    if chunk.shape[0] == 178:

                        # Flatten the 2D chunk array into a 1D array for model input
                        chunk_flattened = chunk.flatten()

                        # Add the flattened chunk to the data list and its corresponding label to the labels list
                        data.append(chunk_flattened)
                        labels.append(label)

                file_count += 1  # Increment file count for the current folder

            # Handle any exceptions during file processing
            except Exception as e:
                print(f"Error processing file {file_path}: {e}")

        print(f"\n\tTotal files read in folder '{folder}': {file_count}")

    # Convert data and labels to NumPy arrays for model input
    return np.array(data), np.array(labels)

# Model Creation
def create_capsule_bilstm_model(input_shape, num_capsules=20, capsule_dim=20, lstm_units=128, dropout_rate=0.5):
    # Define the Capsule Layer function
    CapsuleLayer = create_capsule_layer()

    # Input layer
    input_layer = Input(shape=input_shape)

    # First convolutional layer with L2 regularization and dropout
    x = Conv1D(filters=128, kernel_size=5, activation='relu', padding='same', kernel_regularizer=l2(0.0002))(input_layer)
    x = Conv1D(filters=128, kernel_size=3, activation='relu', padding='same')(x)
    x = Dropout(rate=dropout_rate)(x)
    x = BatchNormalization()(x)
    x = MaxPooling1D(pool_size=2)(x)

    # Capsule layer
    x = CapsuleLayer(num_capsules=num_capsules, capsule_dim=capsule_dim)(x)

    # Bi-directional GRU layers with dropout
    x = Bidirectional(GRU(128, return_sequences=True, kernel_regularizer=l2(0.0002)))(x)
    x = Bidirectional(GRU(64, return_sequences=True))(x)
    x = Dropout(dropout_rate)(x)

    # Flatten and dense layers with L2 regularization and dropout
    x = Flatten()(x)
    x = Dense(units=64, activation='relu', kernel_regularizer=l2(0.0002))(x)
    x = Dropout(rate=dropout_rate)(x)

    # Output layer
    output_layer = Dense(units=3, activation='softmax')(x)

    # Create the model
    model = Model(inputs=input_layer, outputs=output_layer)

    return model

def add_noise(data, noise_factor=0.01):
    # Add Gaussian noise to the data with a specified noise factor
    return data + noise_factor * np.random.normal(size=data.shape)

def analyze_misclassifications(conf_matrix, class_names):
    # Analyze misclassifications based on the confusion matrix
    misclassifications = {}
    for i, row in enumerate(conf_matrix):
        total = sum(row)  # Total samples for the class
        correct = conf_matrix[i, i]  # Correctly classified samples
        incorrect = total - correct  # Misclassified samples
        # Store misclassification stats for the class
        misclassifications[class_names[i]] = {
            "Correct": correct,
            "Incorrect": incorrect,
            "Misclassification Rate": incorrect / total if total > 0 else 0
        }

    # Print the analysis results
    print("\nMisclassification Analysis:")
    for class_name, stats in misclassifications.items():
        print(f"{class_name}: {stats}")

def check_class_symmetry(conf_matrix, class_names):
    # Check symmetry in misclassifications between classes
    print("\nClass Symmetry Analysis:")
    for i in range(len(class_names)):
        for j in range(len(class_names)):
            if i != j:
                # Print misclassification counts for both directions
                print(f"Misclassifications {class_names[i]} → {class_names[j]}: {conf_matrix[i, j]}")
                print(f"Misclassifications {class_names[j]} → {class_names[i]}: {conf_matrix[j, i]}")

def plot_class_distribution(labels, class_names):
    # Plot the distribution of classes in the dataset
    unique, counts = np.unique(labels, return_counts=True)  # Get unique classes and their counts
    plt.bar(class_names, counts)  # Create a bar chart
    plt.title("Class Distribution")  # Set the title
    plt.xlabel("Classes")  # Label the x-axis
    plt.ylabel("Number of Samples")  # Label the y-axis
    plt.show()  # Display the plot

def plot_misclassified_instances(X, y_true, y_pred, class_names, num_instances=5):
    # Plot instances that were misclassified
    misclassified_indices = np.where(y_true != y_pred)[0]  # Find indices of misclassified samples
    print(f"\nTotal Misclassified Instances: {len(misclassified_indices)}")  # Print the total number

    # Plot a limited number of misclassified instances
    for i, idx in enumerate(misclassified_indices[:num_instances]):
        plt.figure(figsize=(10, 4))  # Set the figure size
        plt.plot(X[idx].reshape(-1, 25))  # Plot the EEG signals
        # Set the title with true and predicted class names
        plt.title(f"True: {class_names[y_true[idx]]}, Predicted: {class_names[y_pred[idx]]}")
        plt.xlabel("Time Steps")  # Label the x-axis
        plt.ylabel("EEG Signals")  # Label the y-axis
        plt.show()  # Display the plot

def plot_graphs(history):

    # Define the plots with their corresponding metrics, titles, and labels
    plots = [
        ('accuracy', 'Training Accuracy vs Validation Accuracy', 'Accuracy'),
        ('loss', 'Training Loss vs Validation Loss', 'Loss')
    ]

    # Iterate through the plots and generate each one
    for metric, title, ylabel in plots:
        print(f"\nPlotting {title}...")  # Indicate which plot is being created
        plt.figure(figsize=(10, 6))  # Set the figure size
        plt.plot(history.history[metric], label=f'Training {ylabel}', color='blue')  # Plot the training metric
        plt.plot(history.history[f'val_{metric}'], label=f'Validation {ylabel}', color='orange')  # Plot the validation metric
        plt.title(title)  # Set the title of the plot
        plt.xlabel('Epochs')  # Label the x-axis
        plt.ylabel(ylabel)  # Label the y-axis
        plt.legend()  # Add a legend to differentiate between training and validation
        plt.grid()  # Add a grid for better readability
        plt.show()  # Display the plot

if __name__ == "__main__":
    # Main execution block

    # Set Dataset Path
    base_path = '/kaggle/input/dataset-1-min-20-patients-for-each-case/DataSet Files - 1 min ( 20 patients for each case )'

    # Load and preprocess the dataset
    print("Loading and preprocessing dataset...")
    data, labels = load_and_preprocess_new_dataset(base_path)
    print(f"\nOriginal Data shape: {data.shape}")

    # Reshape data into (samples, timesteps, features) for model compatibility
    print("\nReshaping data into (samples, timesteps, features)...")
    n_samples = data.shape[0]
    data = data.reshape((n_samples, 178, 25))
    print(f"Reshaped Data shape: {data.shape}")

    # Normalize data to ensure consistent scaling
    print("\nNormalizing data...")
    data = data.reshape((n_samples, -1))
    scaler = StandardScaler()
    data = scaler.fit_transform(data)
    data = scaler.fit_transform(data.reshape(-1, 25)).reshape(n_samples, 178, 25)

    # Split the data into training, validation, and testing sets
    print("\nSplitting data into Train, Validation, and Test sets...")
    X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.3, random_state=42, stratify=labels)
    X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42, stratify=y_train)

    # Display the shapes of the split datasets
    print(f"Train set shape: {X_train.shape}")
    print(f"Validation set shape: {X_val.shape}")
    print(f"Test set shape: {X_test.shape}")

    # Add noise to the training data for regularization
    print("\nAdding noise to training data...")
    X_train_noisy = add_noise(X_train)

    # Convert labels to one-hot encoded format for categorical classification
    print("\nOne-hot encoding labels...")
    y_train = to_categorical(y_train, num_classes=3)
    y_val = to_categorical(y_val, num_classes=3)
    y_test = to_categorical(y_test, num_classes=3)

    # Create and compile the Capsule Bi-LSTM model
    print("\nCreating model...")
    input_shape = (178, 25)
    model = create_capsule_bilstm_model(input_shape)
    model.compile(optimizer=Adam(learning_rate=0.0005), loss='categorical_crossentropy', metrics=['accuracy'])
    model.summary()

    # Set up callbacks and compute class weights for balanced training
    print("\nSetting up callbacks and class weights...")
    callbacks = [
        EarlyStopping(monitor='val_loss', patience=30, restore_best_weights=True),
        ModelCheckpoint(filepath='best_model.keras', save_best_only=True)
    ]
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=10, min_lr=1e-7)
    callbacks.append(reduce_lr)

    class_weights = compute_class_weight('balanced', classes=np.unique(labels), y=labels)
    class_weights = {i: w * 1.5 for i, w in enumerate(class_weights)}
    print(f"Class weights: {class_weights}")

    # Train the model with the training data
    print("\nTraining model...")
    history = model.fit(X_train_noisy, y_train, validation_data=(X_val, y_val), epochs=150, batch_size=64, callbacks=callbacks, class_weight=class_weights)

    # Generate predictions and evaluate the model
    print("\nGenerating predictions and metrics...")
    y_pred = model.predict(X_test)
    y_pred_classes = np.argmax(y_pred, axis=1)
    y_true = np.argmax(y_test, axis=1)

    # Compute and display the confusion matrix
    conf_matrix = confusion_matrix(y_true, y_pred_classes)
    print("\nConfusion Matrix:")
    sns.heatmap(conf_matrix, annot=True, fmt="d")
    plt.show()

    # Print the classification report for detailed metrics
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred_classes))

    # Plot training and validation performance graphs
    plot_graphs(history)

    # Analyze misclassifications in the predictions
    print("\nAnalyzing Misclassifications...")
    class_names = ['Normal', 'Focal', 'General']
    analyze_misclassifications(conf_matrix, class_names)

    # Check class symmetry in the confusion matrix
    print("\nChecking Class Symmetry...")
    check_class_symmetry(conf_matrix, class_names)

    # Plot the distribution of classes in the dataset
    print("\nPlotting Class Distribution...")
    plot_class_distribution(labels, class_names)

    # Evaluate the model on the test set
    print("\nFinal Test Evaluation...")
    test_loss, test_accuracy = model.evaluate(X_test, y_test)
    train_accuracy = history.history['accuracy'][-1]
    train_loss = history.history['loss'][-1]

    # Display the training and testing accuracy and loss
    print(f"\nTraining Accuracy: {train_accuracy:.4f}, Train Loss: {train_loss:.4f}")
    print(f"Test Accuracy: {test_accuracy:.4f}, Test Loss: {test_loss:.4f}")

    # Save the trained model to a file
    print("\nSaving Model...")
    model.save('/kaggle/working/Seizure_Classification_Model.h5')

    # Indicate that the process is complete
    print("\nProcess Completed.")

```
